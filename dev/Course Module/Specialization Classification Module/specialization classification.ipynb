```python
import json
import unicodedata
import re
import string
import nltk
from collections import Counter
from itertools import islice, izip
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
import json
import math
import pprint
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
stemmer = PorterStemmer()
lemmatiser = WordNetLemmatizer()
import csv
import pandas as pd
```

```python
import glob

path ='text_files/*.txt'   
files=glob.glob(path)   
```

```python
master_dict={}
for file in files:     
    f=open(file, 'r') 
    print file
    value=f.readlines() 
    value = ''.join(value)
    value =  re.sub("^\d+\s|\s\d+\s|\s\d+$", " ", value)
    value = re.sub(r'\b\d+\b', '', value)
    value=re.sub("\\(.*?\\) ?", '', value);
    value=re.sub("\\[.*?\\] ?", '', value);
    no_digits = []
    result=""
    # Iterate through the string, adding non-numbers to the no_digits list
    for i in value:
        if not i.isdigit():
            no_digits.append(i)
    result = ''.join(no_digits)
    value=result.lower()
    shortword = re.compile(r'\W*\b\w{1,3}\b')
    value=shortword.sub('', value)
    words=re.findall(r'\w+', value)
    new_words=[]
    for w in words:
        tokens_pos = pos_tag(w)
        pos = tokens_pos[0][1]
        if 'V' in pos:
             meta = stemmer.stem(w)
        else:
             meta = lemmatiser.lemmatize(w)
        if not len(meta) < 3:
            new_words.append(str(meta))
    #Generating bigrams
    bigrams=zip(new_words, new_words[1:])
    word_counts = Counter(izip(new_words, islice(new_words, 3, None)))
    master_dict[file]=bigrams
    f.close()  
```

```python
business_vector=master_dict["text_files/business_intelligence.txt"]
ml_vector=master_dict["text_files/data_science.txt"]
hci_vector=master_dict["text_files/ux_design.txt"]
consulting_vector=master_dict["text_files/consulting.txt"]
cyber_vector=master_dict["text_files/cyber_security.txt"]
info_arch_vector=master_dict["text_files/information_architecture.txt"]
mlis_vector=master_dict["text_files/mlis.txt"]
comp_net_vector=master_dict["text_files/computer_networks.txt"]
web_mob_vect=master_dict["text_files/web_mobile_development.txt"]
sde_vector=master_dict["text_files/software_engineering.txt"]
distributed_vector=master_dict["text_files/distributed_systems_cs.txt"]
```

```python
#Generating bigrams for all programs
with open('cur_data.json') as json_data:
    d = json.load(json_data)
    
data_dict = {}

for key, value in d.iteritems():
    value=unicodedata.normalize('NFKD', value).encode('ascii','ignore')
    value =  re.sub("^\d+\s|\s\d+\s|\s\d+$", " ", value)
    value = re.sub(r'\b\d+\b', '', value)
    value=re.sub("\\(.*?\\) ?", '', value);
    value=re.sub("\\[.*?\\] ?", '', value);
    no_digits = []
    result=""
    # Iterate through the string, adding non-numbers to the no_digits list
    for i in value:
        if not i.isdigit():
            no_digits.append(i)
    result = ''.join(no_digits)
    value=result.lower()
    shortword = re.compile(r'\W*\b\w{1,3}\b')
    value=shortword.sub('', value)
    words=re.findall(r'\w+', value)
    new_words=[]
    for w in words:
        tokens_pos = pos_tag(w)
        pos = tokens_pos[0][1]
        if 'V' in pos:
             meta = stemmer.stem(w)
        else:
             meta = lemmatiser.lemmatize(w)
        if not len(meta) < 3:
            new_words.append(str(meta))
        bigrams=zip(new_words, new_words[1:])
    data_dict[key]=bigrams
```

```python
# The following code creates a dictionary with each program id as a key and value as a nested dicitonairy of 
#specializations and their score
program_weights={}
for program, bigrams in data_dict.iteritems():
    d_temp={}
    threshold=len(bigrams)
    
    count=0
    for each in bigrams:
        if each in business_vector:
            count=count+1
    match_percentage=float(count)/float(threshold)*100
    d_temp['business_intelligence']=match_percentage
    program_weights[program]=d_temp
    
    
    
    

    count=0
    for each in bigrams:
        if each in ml_vector:
            count=count+1
    match_percentage=float(count)/float(threshold)*100
    d_temp['data_science_analytics']=match_percentage
    program_weights[program]=d_temp
    
    
    
    
    count=0
    for each in bigrams:
        if each in hci_vector:
            count=count+1
    match_percentage=float(count)/float(threshold)*100
    d_temp['human_center_design_engineering']=match_percentage
    program_weights[program]=d_temp
    
    
    
    
    count=0
    for each in bigrams:
        if each in consulting_vector:
            count=count+1
    match_percentage=float(count)/float(threshold)*100
    d_temp['management_consulting']=match_percentage
    program_weights[program]=d_temp
    
    
    
    
    count=0
    for each in bigrams:
        if each in cyber_vector:
            count=count+1
    match_percentage=float(count)/float(threshold)*100
    d_temp['information_assurance_cyber_security']=match_percentage
    program_weights[program]=d_temp
    
    
    
    
    count=0
    for each in bigrams:
        if each in info_arch_vector:
            count=count+1
    match_percentage=float(count)/float(threshold)*100
    d_temp['information_architecture']=match_percentage
    program_weights[program]=d_temp
    
    
    
    
    count=0
    for each in bigrams:
        if each in mlis_vector:
            count=count+1
    match_percentage=float(count)/float(threshold)*100
    d_temp['library_science']=match_percentage
    program_weights[program]=d_temp
    
    
    
    
    count=0
    for each in bigrams:
        if each in comp_net_vector:
            count=count+1
    match_percentage=float(count)/float(threshold)*100
    d_temp[ 'computer_networks']=match_percentage
    program_weights[program]=d_temp
    
    
    
    
    count=0
    for each in bigrams:
        if each in web_mob_vect:
            count=count+1
    match_percentage=float(count)/float(threshold)*100
    d_temp[ 'web_application_development']=match_percentage
    program_weights[program]=d_temp
    
    
    
    
    count=0
    for each in bigrams:
        if each in sde_vector:
            count=count+1
    match_percentage=float(count)/float(threshold)*100
    d_temp['software_engineering']=match_percentage
    program_weights[program]=d_temp
    
    
    
    
    count=0
    for each in bigrams:
        if each in distributed_vector:
            count=count+1
    match_percentage=float(count)/float(threshold)*100
    d_temp['distributed_systems']=match_percentage
    program_weights[program]=d_temp

    
program_weights
```
