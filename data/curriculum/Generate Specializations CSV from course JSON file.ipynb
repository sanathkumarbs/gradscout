```python
import json
import unicodedata
import re
import string
import nltk
from collections import Counter
from itertools import islice, izip
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
import json
import math
import pprint
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
stemmer = PorterStemmer()
lemmatiser = WordNetLemmatizer()
import csv
import pandas as pd
```

```python
#some of the common;y occuring words that I had removed manually 
unwanted_words=["available","delivery","enrollment", "spring", "summer", "autum", "winter", "online", 
                "next", "offer", "optional"]
```

```python
#Reading the data from cur_data.json
with open('cur_data.json') as json_data:
    d = json.load(json_data)
```

```python
#the following code generates the bigrams for the programs which will firther be used to generate standard vectors 
#of bigrams for each specializations
master_dict = {}

for key, value in d.iteritems():
    value=unicodedata.normalize('NFKD', value).encode('ascii','ignore')
    value =  re.sub("^\d+\s|\s\d+\s|\s\d+$", " ", value)
    value = re.sub(r'\b\d+\b', '', value)
    value=re.sub("\\(.*?\\) ?", '', value);
    value=re.sub("\\[.*?\\] ?", '', value);
    no_digits = []
    result=""
    # Iterate through the string, adding non-numbers to the no_digits list
    for i in value:
        if not i.isdigit():
            no_digits.append(i)
    result = ''.join(no_digits)
    value=result.lower()
    for each in unwanted_words:
        value=re.sub(each,'',value)
    shortword = re.compile(r'\W*\b\w{1,3}\b')
    value=shortword.sub('', value)
    words=re.findall(r'\w+', value)
    new_words=[]
    for w in words:
        tokens_pos = pos_tag(w)
        pos = tokens_pos[0][1]
        if 'V' in pos:
             meta = stemmer.stem(w)
        else:
             meta = lemmatiser.lemmatize(w)
        if not len(meta) < 3:
            new_words.append(str(meta))
    #Generating bigrams
    bigrams=zip(new_words, new_words[1:])
    word_counts = Counter(izip(new_words, islice(new_words, 3, None)))
    master_dict[key]=bigrams
    

#print master_dict['28.htm']
```

```python
#Generating bigrams for all programs
with open('cur_data.json') as json_data:
    d = json.load(json_data)
    
data_dict = {}

for key, value in d.iteritems():
    value=unicodedata.normalize('NFKD', value).encode('ascii','ignore')
    value =  re.sub("^\d+\s|\s\d+\s|\s\d+$", " ", value)
    value = re.sub(r'\b\d+\b', '', value)
    value=re.sub("\\(.*?\\) ?", '', value);
    value=re.sub("\\[.*?\\] ?", '', value);
    no_digits = []
    result=""
    # Iterate through the string, adding non-numbers to the no_digits list
    for i in value:
        if not i.isdigit():
            no_digits.append(i)
    result = ''.join(no_digits)
    value=result.lower()
    for each in unwanted_words:
        value=re.sub(each,'',value)
    shortword = re.compile(r'\W*\b\w{1,3}\b')
    value=shortword.sub('', value)
    words=re.findall(r'\w+', value)
    new_words=[]
    for w in words:
        tokens_pos = pos_tag(w)
        pos = tokens_pos[0][1]
        if 'V' in pos:
             meta = stemmer.stem(w)
        else:
             meta = lemmatiser.lemmatize(w)
        if not len(meta) < 3:
            new_words.append(str(meta))
        bigrams=zip(new_words, new_words[1:])
    data_dict[key]=bigrams
```
